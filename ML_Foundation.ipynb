# Linear algebra

**Solve and visualize:**

- On April 1st, Jill's `Mark 1` design begins generating power 1kj/day.
- On May 1st, her `Mark 2` design begins generating power 4kj/day.

Questions:
1. What day is when Jill's mark 2 design has generated as much total engery as the Mark 1 design?
2. How much total energy have both generated by that day?
3. What would the solutions to (1.) and (2.) be if Mark 2 design generated 1kj of power per day.

```python
import numpy as np
days = np.arange(0, 60)
days
```

# Equation
- **p_1 = 1 * d**
- **p_2 = 4 * (d - 30)**

**Solving for how many days**
- 1 * d = 4 * (d - 30)
- 1d = 4d - 120
- 120 = -1d + 4d
- 120 = 3d
- **d = 40 days**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
# plot for the p_1
p_1 = 1 * days
plt.plot(days, p_1, marker='o', linestyle='-', color='b', label='Y = 1 * d')
# plot for the p_2
# -30 days because it starts late
days_after_30days = days[30:]
p_2 = 4 * (days_after_30days - 30)
plt.plot(days_after_30days, p_2, marker='x', linestyle='-', color='r', label='Y = 4 * (d - 30)')

# Add labels and title
plt.xlabel('Total Days (X)')
plt.ylabel('Total Power (Y)')
plt.title('Total Power vs. Total Days')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
